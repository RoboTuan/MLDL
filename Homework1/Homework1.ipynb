## Import Libraries

```python
import pandas as pd
import numpy as np

from itertools import combinations

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler


from sklearn.datasets import load_wine

from sklearn.metrics import accuracy_score

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.colors import ListedColormap

from sklearn.model_selection import GridSearchCV

```

## Utility functions (some functions that are needed multiple times)

```python
def get_best_parameter(scores, parameter_list):
    """
    Searches for indexes where the score is maximum and returns the corresponding values
    in the list of parameters
    
    """
#     best_param = parameter_list[np.argmax(np.array(scores))]
   
    best_params = [parameter_list[index] for index in np.argwhere(scores == np.max(np.array(scores))).flatten().tolist()]
    print(f"Best parameters: {best_params}")
    return best_params
```

```python
def classify(clf, X_train, y_train, X_test, y_test):
    """
    Recieve data and classifiers, return accuracy on the testing set. If more classifiers are passed returns the best result
    
    """
    y_test_scores = []
    
    for classifier in clf:
        
        
        classifier.fit(X_train, y_train)
        y_predict = classifier.predict(X_test)
        y_test_scores.append(accuracy_score(y_test, y_predict))
    
    return max(y_test_scores)
```

```python
def print_score(scores, type_of_score):
    """
    Print the evaluated scores, their mean and standard deviation.
    """
    print(f"{type_of_score} for each iteration: {scores}")
    mean = scores.mean()
    std = scores.std() * 2
    print(f"{type_of_score} (statistics): {mean:.3f} (+/- {std:.3f})")
```

```python
def plot_decision_boundaries(X, y, clf, figure, ax, title):
    h = .02  # step size in the mesh

    # Create color maps
    cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
    cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    #plt.figure()
    ax.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    ax.set_title(title)

#     plt.show()
```

## Load the Dataset

```python
def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df

data = sklearn_to_df(load_wine())
labels = data['target']
data.drop(columns=['target'], inplace=True)
data.head()
```

```python
data_al_mal = data.drop(columns=data.columns[2:])
data_al_mal.head()
```

## Select the first two attributes for a 2D representation of the image

```python
ax = sns.scatterplot(x="alcohol", y="malic_acid", data=data_al_mal, hue=labels)
```

## Split train_val_test

```python
scaler = StandardScaler()

# Vedere se fare lo scaling separatamente per il test e train test (ma fittando sempre sul training solo)
# data_al_mal = scaler.fit_transform(data_al_mal)

X_train, X_test, y_train, y_test = train_test_split(data_al_mal, labels, test_size=0.3, stratify=labels, random_state=0)

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=2/7, stratify=y_train, random_state=0) #5:2:3
```

## KNeighborsClassifier

### a) Apply K-Nearest Neighbors
### b) Plot the data and the decision boundaries
### c) Evaluate the method on the validation set

```python
K = [1, 3, 5, 7]
scores = []

def do_KNeigh(X_train, X_val, y_train, y_val, K):
    
    # Set up the matplotlib figure
    figure, axes = plt.subplots(2, 2, figsize=(20, 20))
    
    flat_axes = [ax for subaxes in axes for ax in subaxes]

    for i, k in enumerate(K):
        
        clf_KNN = KNeighborsClassifier(n_neighbors=k)
        clf_KNN.fit(X_train, y_train)
        
        y_val_predict = clf_KNN.predict(X_val)
        scores.append(accuracy_score(y_val, y_val_predict))
        
        title = title = "3-Class classification c = %.f, accuracy = %.3f" % (k, round(scores[i], 3))
        
        plot_decision_boundaries(X_train, y_train.to_numpy(), clf_KNN, figure, flat_axes[i], title)
    
    plt.show()


do_KNeigh(X_train, X_val, y_train, y_val, K)

# print(f"The scores are {scores}")
print_score(np.array(scores), "accuracy")
```

### Plot a graph showing how the accuracy on the validation set varies when changing K

```python
ax = sns.lineplot(x=K, y=scores)
ax.set(xlabel='K Neighbours', ylabel='Accuracy')
ax.set_title("Variation of accuracy with respect to K")
plt.show()
```

### Evaluate the model on the test set

```python
bestK = get_best_parameter(scores, K)

KNNs = [KNeighborsClassifier(n_neighbors=parameter) for parameter in bestK]

y_test_score = classify(KNNs, X_train, y_train, X_test, y_test)

y_test_score
```

## Support Vector Machines

### a) Train a linear SVM on the training set
### b) Plot the data and the decision boundaries
### c) Evaluate the method on the validation set

```python
C = [0.001, 0.01, 0.1, 1, 10, 100,1000]

scores = []

def doSvc_linear(X_train, X_val, y_train, y_val, C):
    
    # Set up the matplotlib figure
    figure, axes = plt.subplots(3, 3, figsize=(20, 20))
    
    flat_axes = [ax for subaxes in axes for ax in subaxes]
    
    for i, c in enumerate(C):
        
        clf_SVC = SVC(random_state=0, C=c, kernel = 'linear')
        clf_SVC.fit(X_train, y_train)
        
        y_val_predict = clf_SVC.predict(X_val)
        scores.append(accuracy_score(y_val, y_val_predict))
        
        title = "3-Class classification c = %.f, accuracy = %.3f" % (c, round(scores[i], 3))
        
        plot_decision_boundaries(X_train, y_train.to_numpy(), clf_SVC, figure, flat_axes[i], title)
    
    figure.delaxes(axes[2,1])
    figure.delaxes(axes[2,2])
    plt.show()
        
        
doSvc_linear(X_train, X_val, y_train, y_val, C)

# print(f"The scores are {scores}")
print_score(np.array(scores), "accuracy")
```

### Plot a graph showing how the accuracy on the validation set varies when changing C

```python
ax = sns.lineplot(x=list(np.log10(c) for c in C), y=scores)
ax.set(xlabel='Regularization parameter C', ylabel='Accuracy')
ax.set_title("Variation of accuracy with respect to C")
plt.show()
```

### Use the best value of C and evaluate the model on the test set.

```python
bestC = get_best_parameter(scores, C)

# bestC = 1

clf_SVCs = [SVC(random_state=0, C=parameter, kernel = 'linear') for parameter in bestC]

classify(clf_SVCs,  X_train, y_train, X_test, y_test)
```

## Kernel SVM (RBF)

### Repeat point 8) (train, plot, etc..), but this time use an RBF kernel

```python
C = [0.001, 0.01, 0.1, 1, 10, 100,1000]

scores = []

def doSvc_RBF(X_train, X_val, y_train, y_val, C):
    
    # Set up the matplotlib figure
    figure, axes = plt.subplots(3, 3, figsize=(20, 20))
    
    flat_axes = [ax for subaxes in axes for ax in subaxes]
    
    for i, c in enumerate(C):
        
        clf_SVC_RBF = SVC(random_state=0, C=c, kernel = 'rbf')
        clf_SVC_RBF.fit(X_train, y_train)
        
        y_val_predict = clf_SVC_RBF.predict(X_val)
        scores.append(accuracy_score(y_val, y_val_predict))
        
        title = "3-Class classification c = %.f, accuracy = %.3f" % (c, round(scores[i], 3))
        
        plot_decision_boundaries(X_train, y_train.to_numpy(), clf_SVC_RBF, figure, flat_axes[i], title)

    plt.show()
        
        
doSvc_linear(X_train, X_val, y_train, y_val, C)

print(f"The scores are {scores}")
```

```python
ax = sns.lineplot(x=list(str(c) for c in C), y=scores)
ax.set(xlabel='Regularization parameter C', ylabel='Accuracy')
ax.set_title("Variation of accuracy with respect to C")
plt.show()
```

### Evaluate the best C on the test set

```python
bestC = get_best_parameter(scores, C)

clf_SVC_RBFs = [SVC(random_state=0, C=index, kernel = 'rbf') for index in bestC]

classify(clf_SVC_RBFs,  X_train, y_train, X_test, y_test)
```

### Perform a grid search of the best parameters for an RBF kernel

```python
param_grid = {'gamma':[0.01, 0.1, 1, 10, 100], 'C':[0.01, 0.1, 1, 10, 100]}

scores = []

def GridSearch(parameters):
    
    figure, axes = plt.subplots(9, 3, figsize=(20, 40))
    
    flat_axes = [ax for subaxes in axes for ax in subaxes]
    
    i = 0
    
    for gamma in parameters['gamma']:
        for c in parameters['C']:
            SVM_model = SVC(random_state=0, kernel='rbf', gamma=gamma, C=c)
            SVM_model.fit(X_train, y_train)
            y_val_predict = SVM_model.predict(X_val)
            
            y_train_predict = SVM_model.predict(X_train)
            
            # Can add the training accuracy into scores to see the overfitting at the last pictures
            #scores.append((gamma, c, accuracy_score(y_val, y_val_predict), accuracy_score(y_train, y_train_predict)))
            scores.append((gamma, c, accuracy_score(y_val, y_val_predict)))
            
            title = "3-Class classification c = %.f, gamma = %.f, accuracy = %.3f" % (c, gamma, round(scores[i][2], 3))
        
            plot_decision_boundaries(X_train, y_train.to_numpy(), SVM_model, figure, flat_axes[i], title)
            
            i += 1
            
    figure.delaxes(axes[8,1])
    figure.delaxes(axes[8,2])
    plt.show()
            
        
    
    
GridSearch(param_grid)

index = np.argmax([i[2] for i in scores])
best_gamma = scores[index][0]
best_c = scores[index][1]


print(f"Best parameters: c={best_c}, gamma={best_gamma}")
print(f"Best accuracy on validation: {scores[index][2]}")

clf_SVC_RBF = SVC(random_state=0, kernel='rbf', gamma=best_gamma, C=best_c)
clf_SVC_RBF.fit(X_train, y_train)
y_val_predict = clf_SVC_RBF.predict(X_test)
score = accuracy_score(y_test, y_val_predict)

print(f"Accuracy on the test set = {score}")
```

## Try different pairs of attributes

```python
attributes = list(data.columns.values)
pair_attributes = list(combinations(attributes, 2))
pair_attributes.remove(("alcohol", "malic_acid"))

for pair in pair_attributes:
    
    new_X_train, new_x_test, new_y_train, new_y_test = train_test_split(data[list(pair)], labels, test_size=0.3, stratify=labels, random_state=0)

    new_X_train = scaler.fit_transform(new_X_train)
    new_x_test = scaler.transform(new_x_test)
    
    # Hyper parameters tuning
    SVM_model = SVC(random_state=0, kernel='rbf')
    param_grid = {'gamma':[0.01, 0.1, 1, 10, 100], 'C':[0.01, 0.1, 1, 10, 100]}
    gridSearch = GridSearchCV(SVM_model, param_grid, scoring='accuracy', cv=5, n_jobs=-1)
    gridSearch.fit(new_X_train, new_y_train)
    
    # Traning best model and testing
    clf_SVC_RBF = gridSearch.best_estimator_
    clf_SVC_RBF.fit(new_X_train, new_y_train)
    y_test_predict = clf_SVC_RBF.predict(new_x_test)
    score = accuracy_score(new_y_test, y_test_predict)
    
    if score > 0.88:        
        print(f"Classifier with attributes: {pair}")
        print(f"Best parameters of the classifier: {gridSearch.best_params_}")
        print(f"Accuracy on the test set = {score}")
        
        
        # Set up the matplotlib figure
        figure, axes = plt.subplots(1, 2, figsize=(20, 10))    
        flat_axes = [ax for ax in axes]
        
        title = title = "Train"
        plot_decision_boundaries(new_X_train, new_y_train.to_numpy(), clf_SVC_RBF, figure, flat_axes[0], title)
        
        title = title = "Test"
        plot_decision_boundaries(new_x_test, new_y_test.to_numpy(), clf_SVC_RBF, figure, flat_axes[1], title)
        plt.show()
        print()
        print()
        print()
    
```
